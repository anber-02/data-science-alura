### **Normalizando archivos JSON**

```python
'''
    Otras soluciones para normalizar los datos es la siguiente
    # aquí indicamos al cargar los datos donde esta la columna que esta anidada
    
    pd.json_normalize(datos_pacientes_2, record_path=['Pacientes'])
    
    como se pude observar se eliminan los campos ["investigación", "Año"]
    para poder agregarlos existe el parámetro Meta
    
    pd.json_normalize(datos_pacientes_2, record_path=['Pacientes'], meta=['Investigación', 'Año'])
    
'''
df_normalizado = pd.json_normalize(datos_pacientes_2['Pacientes'])
df_normalizado
# pd.json_normalize(df_normalizado['D'])
# datos_pacientes_2_normalized = pd.read_json('./resources/data/pacientes_2.json')
# datos_pacientes_2_normalized
```
